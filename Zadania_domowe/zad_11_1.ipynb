```python
import pandas as pd
```

```python
csv_url = 'https://github.com/piotrgradzinski/python_20200328/raw/master/data_science_pgg/data/matches.csv'
data = pd.read_csv(csv_url, parse_dates=['date'])
data
```

```python
team = 'Śląsk Wrocław'
year = 2013
home_scores = data[
    
        (data.home_team == team) & (data.date.dt.year.isin(range(year, year + 1)))]
home_scores
```

```python
away_scores = data[
    
        (data.away_team == team) & (data.date.dt.year.isin(range(year, year + 1)))]
away_scores
```

```python
scores1 = home_scores['home_team_goal'].sum()
```

```python
scores2 = away_scores['away_team_goal'].sum()
```

```python
total = scores1 + scores2
total
```
